- Youtube Tutorial 9 [WordNet - Natural Language Processing With Python and NLTK p.10](https://www.youtube.com/watch?v=T68P5-8tM-Y&list=PLQVvvaa0QuDf2JswnfiGkliBInZnIC4HL&index=10)
- WordNet: Largest/richest Corpus in NLTK

```python
from nltk.corpus import wordnet

syns = wordnet.synsets("program")

# Synsets
print(syns[0].name())

# just the word
print(syns[0].lemmas()[0].name())

# Definition
print(syns[0].definition())

# Example
print(syns[0].examples())

# Synonyms and antonyms
synonyms = []
antonyms = []

for syn in wordnet.synsets('good'):
    print(syn)
    for l in syn.lemmas():
        synonyms.append(l.name())
    if l.antonyms():
        antonyms.append(l.antonyms()[0].name())
        
# Print out the lists, and 
# remove the duplicates by going through sets
print('Synonyms: ', set(synonyms))
print('Antonyms: ', set(antonyms))


# Compare semantic similarity of words
w1 = wordnet.synset('ship.n.01') # .n.01 - noun, first one
w2 = wordnet.synset('boat.n.01')

print(w1.wup_similarity(w2)) # returns 0-1 % of similarity

w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('car.n.01')
print(w1.wup_similarity(w2))

w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('cat.n.01')
print(w1.wup_similarity(w2))

w1 = wordnet.synset('cat.n.01')
w2 = wordnet.synset('dog.n.01')
print(w1.wup_similarity(w2))

w1 = wordnet.synset('human.n.01')
w2 = wordnet.synset('alien.n.01')
print(w1.wup_similarity(w2))
```
